# **ColumnTransformer**

**So far, One-Hot Encoding has been applied manually using Pandas and NumPy.**
**While this helps understand the concept, it becomes difficult to manage when:**

**Different columns require different preprocessing**

**Train and test data must be handled consistently**

**The preprocessing steps need to be reused or deployed**

**To solve these problems, Scikit-learn provides ColumnTransformer, which allows applying different transformations to different columns in a single, unified pipeline.**

**--Limitations of Manual Encoding**

1) Manual use of pd.get_dummies() and np.hstack():
2) Risks data leakage
3) Requires repeated code for train and test data
4) Does not scale well for production
5) Hard to track feature order and column alignment

**What ColumnTransformer Solves**

ColumnTransformer enables feature-wise preprocessing, where:

Categorical columns can be one-hot encoded

Numerical columns can be passed through or scaled

All preprocessing is fit only on training data

The same transformations are applied to test data automatically

here we import **SimpleImputer** its work for

**Purpose**: Fill in missing values in your dataset so your machine learning models don‚Äôt break.

**üîπ How it Works**

Looks for missing values (NaN, or a specific placeholder)

Replaces them with a strategy value:

mean ‚Üí replace with the average (numeric)

median ‚Üí replace with the median (numeric)

most_frequent ‚Üí replace with the most common value (categorical)

constant ‚Üí replace with a fixed value you provide

**Example**
from sklearn.impute import SimpleImputer

For numeric columns: replace NaN with mean\
**imputer = SimpleImputer(strategy='mean')**

Fit on training data\
**imputer.fit(X_train[['age', 'salary']])**

Transform training and test data\
**X_train[['age', 'salary']] = imputer.transform(X_train[['age', 'salary']])**\
**X_test[['age', 'salary']] = imputer.transform(X_test[['age', 'salary']])**

#  Using ColumnTransformer

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3',OneHotEncoder(sparse_output=False,drop='first'),['gender','city'])
],remainder='passthrough') 
#remainder='passthrough' tells the transformer:
#‚ÄúFor any column I didn‚Äôt explicitly list, just leave it as it is and include it in the output.‚Äù
#They are passed through unchanged.
```

Components:

SimpleImputer ‚Üí handles missing numeric values.

OrdinalEncoder ‚Üí converts ordered categories (Mild < Strong) into integers.

OneHotEncoder ‚Üí converts categorical variables into dummy variables (with drop='first' to avoid multicollinearity).

remainder='passthrough' ‚Üí all columns not mentioned are left as-is.
