 # Working with Dates

```python
# Converting to datetime datatype
data1['date'] = pd.to_datetime(data1['date'])
```

```python
data1.info()
```

# 1. Extract year

```python
data1['year'] = data1['date'].dt.year

data1.sample(5)
```

# 2. Extract Month

```python

data1['month'] = data1['date'].dt.month

data1.head()
```

```python
data1['month_name'] = data1['date'].dt.month_name()

data1.head()
```

# Extract Days

```python

data1['day'] = data1['date'].dt.day

data1.head()
```

```python
# day of week
data1['day_name'] = data1['date'].dt.dayofweek

data1.head()
```

```python
# day of week - name

data1['day_name'] = data1['date'].dt.day_name()

data1.drop(columns=['product_id','city_id','orders']).head()
```

```python
# is weekend?

data1['is_weekend'] = np.where(data1['day_name'].isin(['Sunday', 'Saturday']), 1,0)
```

```python
data1.head()
```

```python
# Extract week of the year
data1['date_week'] = data1['date'].dt.isocalendar().week
data1.head()
```

```python
data1['quarter'] = data1['date'].dt.quarter
data1.head()
```

```python
# Extract Semester
data1['semester'] = np.where(data1['quarter'].isin([1,2]), 1, 2)
data1.head()
```

**Extract Time elapsed between dates**

```python
import datetime

today = datetime.datetime.today()

today
```

```python
today - data1['date']
```

```python
(today - data1['date']).dt.days
```
