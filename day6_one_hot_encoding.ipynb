# **1. OneHotEncoding using Pandas**

```python
pd.get_dummies(df,columns=['fuel','owner'])
```

By default, pd.get_dummies() may return boolean (True / False) values depending on the Pandas version and settings.

The **dtype=int** parameter is used to force the encoded columns to be numeric (0 and 1) instead of boolean.

```python
pd.get_dummies(df,columns=['fuel','owner'],dtype=int)
```

# **2.  K-1 OneHotEncoding**

If a categorical feature has K unique categories, standard One-Hot Encoding creates K binary columns.

K-1 encoding drops one category and keeps only K-1 columns.

When all K dummy variables are included, one column can be perfectly predicted from the others.

This creates multicollinearity, which causes instability in some machine learning models.

Dropping one category does NOT remove information.

The dropped category becomes the reference category, inferred when all other encoded values are zero.

```python
pd.get_dummies(df,columns=['fuel','owner'],drop_first=True,dtype=int)
```

# **3. OneHotEncoding using Sklearn**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=2)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop='first',sparse_output=False,dtype=np.int32)
```

```python
X_train_new = ohe.fit_transform(X_train[['fuel','owner']])
X_test_new = ohe.transform(X_test[['fuel','owner']])
```

```python
X_train_new.shape
```

After applying One-Hot Encoding to the categorical features (fuel and owner), the resulting output (X_train_new) is a NumPy array containing only the encoded categorical columns.

The original numerical features (km_driven) and unencoded features (brand) are not included in this array.

np.hstack() is used to horizontally combine:

the original columns (brand, km_driven) and

the one-hot encoded categorical features

This creates a single feature matrix where each row contains all required input features for the machine learning model.

```python
np.hstack((X_train[['brand','km_driven']].values,X_train_new))
```

The brand column is still categorical (string values) in this combined array.
Machine learning models cannot work with string features, so brand must also be encoded (One-Hot / Top Categories / Target Encoding) before training a model.

# **OneHotEncoding with Top Categories**

**Rare categories do not have enough data** 

**The model cannot learn reliable patterns from them**

**Grouping them reduces variance**

**" This is a bias–variance tradeoff win. "**

Some categorical features (like brand) contain many unique values.
Applying standard One-Hot Encoding directly on such features can:

Create too many columns

Increase sparsity

Lead to overfitting and poor generalization

To handle this, One-Hot Encoding with Top Categories is used.

```python
counts = df['brand'].value_counts()
counts
```

```python
df['brand'].nunique()
threshold = 100
```

Categories appearing 100 times or fewer are treated as rare.

**The threshold value is not fixed and should be chosen based on data size**

**This technique should be applied after train-test split in production systems**

```python
repl = counts[counts <= threshold].index
```

```python
pd.get_dummies(df['brand'].replace(repl, 'uncommon'),dtype=int).sample(5)
```
